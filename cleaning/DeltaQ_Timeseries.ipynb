## Calculate $\Delta Q$ features

```python
def crop_data(id,cycle):
    # isolate discharging data
    data = np.array([[Q, V] for Q, V in zip(bat_dict[id]["cycles"][cycle]["Qd"], bat_dict[id]["cycles"][cycle]["V"]) if Q > 1e-4])
    cropped = [data[0]] 
    for i, datapoint in enumerate(data[1:]):
        if datapoint[1] > cropped[-1][1]:
            if datapoint[1] < 2.2: # identify where discharging ends
                break
            else: # this is an anomalous increase in voltage; remove so voltage is decreasing
                pass
        else:
            cropped.append(datapoint) # add datapoints one by one
    cropped = np.array(cropped)
    x = np.flip(cropped[:, 1]) # voltage
    y = np.flip(cropped[:, 0]) # discharge capacity
    return x,y
```

```python
xs = np.linspace(2,3.5,1000)
def interpolate_spline_cycle(id,cycle):
    x,y = crop_data(id,cycle)

    cs = UnivariateSpline(x,y,s=0.005)
    return cs

def get_var(id,cs_1,cs_2):
    return np.var(cs_2(xs) - cs_1(xs))

def get_min(id,cs_1,cs_2):
    return np.min(cs_2(xs) - cs_1(xs))

def get_mean(id,cs_1,cs_2):
    return np.mean(cs_2(xs) - cs_1(xs))

def get_skew(id,cs_1,cs_2):
    return scipy.stats.skew(cs_2(xs) - cs_1(xs))

def get_kurt(id,cs_1,cs_2):
    return scipy.stats.kurtosis(cs_2(xs) - cs_1(xs))
```

```python
x = np.arange(4,102,2)
```

```python
for id in bat_dict.keys():
    logVars = []
    mins = []
    means = []
    skews = []
    kurts = []
    cs_1 = interpolate_spline_cycle(id,"2")
    for j in x:
        cs_2 = interpolate_spline_cycle(id,str(j))
        var = get_var(id,cs_1,cs_2)
        logVars.append(var)
        min = get_min(id,cs_1,cs_2)
        mins.append(min)
        mean = get_mean(id,cs_1,cs_2)
        means.append(mean)
        skew = get_skew(id,cs_1,cs_2)
        skews.append(skew)
        kurt = get_kurt(id,cs_1,cs_2)
        kurts.append(kurt)
    feature_dict[id]["DeltaQ_logVars"] = np.array(logVars)
    feature_dict[id]["DeltaQ_mins"] = np.array(mins)
    feature_dict[id]["DeltaQ_means"] = np.array(means)
    feature_dict[id]["DeltaQ_skews"] = np.array(skews)
    feature_dict[id]["DeltaQ_kurts"] = np.array(kurts)
```

## $\Delta Q_{100-10}(V)$ only

```python
feature_dict = {}
for id in bat_dict.keys():
    feature_dict[id] = {"cycle_life":bat_dict[id]["cycle_life"][0][0]}
```

```python
def crop_data(id,cycle):
    # isolate discharging data
    cropped = np.array([[Q, V] for Q, V in zip(bat_dict[id]["cycles"][cycle]["Qd"], bat_dict[id]["cycles"][cycle]["V"]) if Q > 1e-5])
    for i, datapoint in enumerate(cropped):
        if datapoint[1] < cropped[i+1, 1]:
            if datapoint[1] < 2.2: # identify where discharging ends
                end = i+1
                break
            else: # this is an anomalous increase in voltage; remove so voltage is decreasing
                cropped[i+1, 1] = cropped[i, 1]
    cropped = cropped[:end]    
    x = np.flip(cropped[:, 1])
    y = np.flip(cropped[:, 0])
    return x,y
```

```python
def get_deltaQ(id):
    x_100,y_100 = crop_data(id,"100")
    x_10,y_10 = crop_data(id,"10")
    xs = np.linspace(2,3.5,1000)
    cs_100 = UnivariateSpline(x_100,y_100,s=0.001)
    cs_10 = UnivariateSpline(x_10,y_10,s=0.001)
    return cs_100(xs) - cs_10(xs)
```

```python
for id in bat_dict.keys():
    deltaQ = get_deltaQ(id)
    feature_dict[id]["DeltaQ"] = np.array(deltaQ)
```
