## Calculate $\Delta Q$ features

```python
def crop_data(id,cycle):
    # isolate discharging data
    data = np.array([[Q, V] for Q, V in zip(bat_dict[id]["cycles"][cycle]["Qd"], bat_dict[id]["cycles"][cycle]["V"]) if Q > 1e-4])
    cropped = [data[0]] 
    for i, datapoint in enumerate(data[1:]):
        if datapoint[1] >= cropped[-1][1]: # check if voltage is strictly decreasing
            if datapoint[1] < 2.05: # this indicates the end of discharging; cut off series
                break
            else: # this indicates an anomalous jump in voltage; remove but continue iterating
                pass
        else:
            cropped.append(datapoint) # add datapoints one by one
    cropped = np.array(cropped)
    x = np.flip(cropped[:, 1]) # voltage
    y = np.flip(cropped[:, 0]) # discharge capacity
    return x,y
```

```python
def crop_data_temp(id,cycle):
    # isolate discharging data
    data = np.array([[T, V] for T, Q, V in zip(bat_dict[id]["cycles"][cycle]["T"], bat_dict[id]["cycles"][cycle]["Qd"], bat_dict[id]["cycles"][cycle]["V"]) if Q > 1e-4])
    cropped = [data[0]] 
    for i, datapoint in enumerate(data[1:]):
        if datapoint[1] >= cropped[-1][1]: # check if voltage is strictly decreasing
            if datapoint[1] < 2.05: # this indicates the end of discharging; cut off series
                break
            else: # this indicates an anomalous jump in voltage; remove but continue iterating
                pass
        else:
            cropped.append(datapoint) # add datapoints one by one
    cropped = np.array(cropped)
    x = np.flip(cropped[:, 1]) # voltage
    y = np.flip(cropped[:, 0]) # temperature
    return x,y
```

```python
xs = np.linspace(2,3.5,1000)
def interpolate_spline_cycle(id,cycle):
    x,y = crop_data(id,cycle)

    cs = UnivariateSpline(x,y,s=0.001,ext=3) # extrapolation mode = constant
    return cs

def interpolate_spline_temp(id,cycle):
    x,y = crop_data_temp(id,cycle)
    cs = UnivariateSpline(x,y,s=0.001,ext=3) # extrapolation mode = constant
    return cs

def get_var(id,cs_1,cs_2):
    return np.log10(np.var(cs_2(xs) - cs_1(xs)))

def get_min(id,cs_1,cs_2):
    return np.min(cs_2(xs) - cs_1(xs))

def get_mean(id,cs_1,cs_2):
    return np.mean(cs_2(xs) - cs_1(xs))

def get_skew(id,cs_1,cs_2):
    return scipy.stats.skew(cs_2(xs) - cs_1(xs))

def get_kurt(id,cs_1,cs_2):
    return scipy.stats.kurtosis(cs_2(xs) - cs_1(xs))
```

```python
feature_dict = {}
for id in bat_dict.keys():
    feature_dict[id] = {"cycle_life":bat_dict[id]["cycle_life"][0][0]}
```

```python
x = np.arange(3,101,1)
```

```python
for id in bat_dict.keys():
    logVars = []
    mins = []
    means = []
    temp_maxs = []
    temp_vars = []
    deltaIRs = []
    #skews = []
    #kurts = []
    cs_1 = interpolate_spline_cycle(id,"2")
    ts_1 = interpolate_spline_temp(id,"2")
    #cs_1 = interpolate_spline_cycle(id,"10")
    for j in x:
        if id=="b1c0" and j==11: # treat anomalous data point where capacity starts at 1.5 instead of 1.1
            cs_2 = interpolate_spline_cycle(id,str(j-1))
            ts_2 = interpolate_spline_temp(id,str(j-1))
        else:
            cs_2 = interpolate_spline_cycle(id,str(j))
            ts_2 = interpolate_spline_temp(id,str(j))
        var = get_var(id,cs_1,cs_2)
        logVars.append(var)

        min = get_min(id,cs_1,cs_2)
        mins.append(min)

        mean = get_mean(id,cs_1,cs_2)
        means.append(mean)

        temp_max = bat_dict[id]["summary"]["Tmax"][j]-bat_dict[id]["summary"]["Tmax"][2]
        temp_maxs.append(temp_max)

        temp_var = get_var(id,ts_1,ts_2)
        temp_vars.append(temp_var)

        deltaIR = bat_dict[id]["summary"]["IR"][j]-bat_dict[id]["summary"]["IR"][2]
        deltaIRs.append(deltaIR)
        #skew = get_skew(id,cs_1,cs_2)
        #skews.append(skew)
        #kurt = get_kurt(id,cs_1,cs_2)
        #kurts.append(kurt)
    feature_dict[id]["DeltaQ_logVars"] = np.array(logVars)
    feature_dict[id]["DeltaQ_mins"] = np.array(mins)
    feature_dict[id]["DeltaQ_means"] = np.array(means)
    feature_dict[id]["DeltaT_max"] = np.array(temp_maxs)
    feature_dict[id]["DeltaT_logVars"] = np.array(temp_vars)
    feature_dict[id]["DeltaIR"] = np.array(deltaIRs)
    #feature_dict[id]["DeltaQ_skews"] = np.array(skews)
    #feature_dict[id]["DeltaQ_kurts"] = np.array(kurts)
```

## $\Delta Q_{100-10}(V)$ only

```python
feature_dict = {}
for id in bat_dict.keys():
    feature_dict[id] = {"cycle_life":bat_dict[id]["cycle_life"][0][0]}
```

```python
def crop_data(id,cycle):
    # isolate discharging data
    cropped = np.array([[Q, V] for Q, V in zip(bat_dict[id]["cycles"][cycle]["Qd"], bat_dict[id]["cycles"][cycle]["V"]) if Q > 1e-5])
    for i, datapoint in enumerate(cropped):
        if datapoint[1] < cropped[i+1, 1]:
            if datapoint[1] < 2.2: # identify where discharging ends
                end = i+1
                break
            else: # this is an anomalous increase in voltage; remove so voltage is decreasing
                cropped[i+1, 1] = cropped[i, 1]
    cropped = cropped[:end]    
    x = np.flip(cropped[:, 1])
    y = np.flip(cropped[:, 0])
    return x,y
```

```python
def get_deltaQ(id):
    x_100,y_100 = crop_data(id,"100")
    x_10,y_10 = crop_data(id,"10")
    xs = np.linspace(2,3.5,1000)
    cs_100 = UnivariateSpline(x_100,y_100,s=0.001)
    cs_10 = UnivariateSpline(x_10,y_10,s=0.001)
    return cs_100(xs) - cs_10(xs)
```

```python
for id in bat_dict.keys():
    deltaQ = get_deltaQ(id)
    feature_dict[id]["DeltaQ"] = np.array(deltaQ)
```

## Extract charge policy

```python
# get most used charge rate for each battery
charge_info = {}
for id in bat_dict.keys():
    str = bat_dict[id]["charge_policy"]
    str_split = str.split('C')
    c1 = float(str_split[0])
    c2 = str.split('-')[1]
    if 'C' in c2:
        c2 = float(c2.split('C')[0])
    else:
        c2 = float(c2)
    percent = float(str_split[1].split('%')[0][1:])/100
    #if percent >= 50:
    #    charge_info[id] = c1
    #else:
    #    charge_info[id] = c2
    charge_info[id] = (1/c1)*percent + (1/c2)*(1-percent)
```
