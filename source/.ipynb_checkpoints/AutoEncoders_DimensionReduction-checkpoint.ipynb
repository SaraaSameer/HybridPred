```python
class AutoEncoders(Model):
    def __init__(self, output_units):
        super().__init__()
        self.encoder = Sequential([
            Dense(32, activation="relu"),
            Dense(16, activation="relu"),
            Dense(20, activation="relu")    #bottleneck layer -- no. of features
        ])
        self.decoder = Sequential([
            Dense(16, activation="relu"),
            Dense(32, activation="relu"),
            Dense(output_units, activation="sigmoid")
        ])

    def call(self, inputs):
        encoded = self.encoder(inputs)
        decoded = self.decoder(encoded)
        return decoded
```

### **Autoencoders**

```python

# # Assuming you have a 2D array `data` where each row represents a battery with cycle life and DeltaQ_logVars
# data = np.array([
#     [1852.0, [1.075710604433867e-06, 2.3275843758453748e-05]],
#     [2160.0, [1.3099076808194432e-05, 4.548769416167956e-05]],
#     [2237.0, [1.684583961631894e-05, 4.260115928846763e-05]]
#     # Add more rows as needed
# ])

data = train_2d.values

# Separate the cycle life and DeltaQ_logVars
cycle_life = data[:, 0]
deltaq_logvars = np.array(data[:, 1].tolist())

# Scale the input data
scaler = MinMaxScaler()
deltaq_logvars_scaled = scaler.fit_transform(deltaq_logvars)

# Define the autoencoder model
autoencoder = Sequential([
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(20, activation='relu'),  # Set the desired bottleneck layer size (e.g., 8 features)
])

# Add decoder layers to reconstruct the input
autoencoder.add(Dense(16, activation='relu'))
autoencoder.add(Dense(32, activation='relu'))
autoencoder.add(Dense(deltaq_logvars_scaled.shape[1], activation='sigmoid'))  # Output layer with the same number of features as input

# Compile and train the autoencoder
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(deltaq_logvars_scaled, deltaq_logvars_scaled, epochs=20, batch_size=32)

# Obtain the compressed array
compressed_array = autoencoder.predict(deltaq_logvars_scaled)

# Get the indices of selected features
selected_feature_indices = np.argsort(np.std(compressed_array, axis=0))[-20:]  #number of features

# Print the indices
print("Indices of selected features:", selected_feature_indices)
```

```python
def loss(ground_truth, predictions):
    diff = np.sqrt(np.mean((10**ground_truth - 10**predictions)**2))
    return diff

my_scorer = make_scorer(loss, greater_is_better=False)

regressor = ElasticNet()
param_grid = {
    'alpha': np.logspace(-6, 0, 10000),  # Range of alpha values to explore
    'l1_ratio': np.logspace(-0.01, 0, 100000)
}
print(y_train)
folds = KFold(n_splits=4, shuffle=True, random_state=42)
model = RandomizedSearchCV(
    estimator=regressor,
    n_iter=20000,
    cv=folds,
    param_distributions=param_grid,
    scoring=my_scorer,
    random_state=42,
    verbose=2,
    return_train_score=True,
    n_jobs=-1
)
print(model)
model.fit(x_train, (y_train))
```

```python
regressor = ElasticNet(l1_ratio=model.best_params_["l1_ratio"], alpha=model.best_params_['alpha'])
regressor.fit(x_train,y_train)
```
